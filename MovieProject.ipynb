```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data=pd.read_csv(r"C:\Users\Alekhya Reddy\Desktop\airisData\MoviesProject\input\grossfilms.txt")
data.info()
```

```python
country=data['country'].unique()
gross1=np.array(data.loc[data['rank']==1].gross)
gross2=np.array(data.loc[data['rank']==2].gross)
gross3=np.array(data.loc[data['rank']==3].gross)
x=np.arange(len(country))
width=0.25
fig,ax=plt.subplots()
bar1=ax.bar(x,gross1,width,label='RANK-1')
bar2=ax.bar(x+width,gross2,width,label='RANK-2')
bar3=ax.bar(x+width*2,gross3,width,label='RANK-3')
ax.set_xticks(x)
ax.set_xticklabels(country)
ax.legend()
plt.xticks(rotation=90)
plt.gcf().set_size_inches(50,30)
#plt.figure(figsize=(20,10
plt.title("Top 3 gross films in each country\n", fontdict={'fontsize': 50, 'fontweight' : 10, 'color' : 'Black'})

plt.xlabel("Country", fontdict={'fontsize': 40, 'fontweight' : 20, 'color' : 'Brown'})
plt.ylabel("Gross", fontdict={'fontsize': 40, 'fontweight' : 20, 'color' : 'Brown'})
plt.savefig(r"C:\Users\Alekhya Reddy\Desktop\airisData\MoviesProject\OutputImages\TopGrossFilms.jpg")
plt.show()
```

```python
data1=pd.read_csv(r"C:\Users\Alekhya Reddy\Desktop\airisData\MoviesProject\input\grossYear.txt")
data1.info()
```

```python
year = data1['Year']
gross = data1['Total_gross']

plt.plot(year, gross,marker='*',markersize=4)

plt.title("Gross over Years\n", fontdict={'fontsize': 20, 'fontweight' : 5, 'color' : 'Green'})

# Labeling Axes
plt.xlabel("Year", fontdict={'fontsize': 12, 'fontweight' : 5, 'color' : 'Brown'})
plt.ylabel("Total Gross", fontdict={'fontsize': 12, 'fontweight' : 5, 'color' : 'Brown'} )
plt.gcf().set_size_inches(15,5)
plt.savefig(r"C:\Users\Alekhya Reddy\Desktop\airisData\MoviesProject\OutputImages\grossOverYears.jpg")
plt.show()
```

```python
data2=pd.read_csv(r"C:\Users\Alekhya Reddy\Desktop\airisData\MoviesProject\input\review.txt")
data2.info()
```

```python
ax=sns.countplot(y='genre',data=data2)
ax.bar_label(ax.containers[0])
plt.gcf().set_size_inches(20,10)
plt.xlim([0,2300])
plt.xticks(range(0,2300,100))
plt.xticks(rotation=90)
plt.title("Genre wise count", fontdict={'fontsize': 30, 'fontweight' : 10, 'color' : 'Green'})

# Labeling Axes
plt.xlabel("Count", fontdict={'fontsize': 20, 'fontweight' : 10, 'color' : 'Brown'})
plt.ylabel("Genre", fontdict={'fontsize': 20, 'fontweight' : 10, 'color' : 'Brown'} )
plt.savefig(r"C:\Users\Alekhya Reddy\Desktop\airisData\MoviesProject\OutputImages\genreWiseCount.jpg")
plt.show()
```
